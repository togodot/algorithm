# 02 Filtering & Sorting

### 식당데이터 : https://github.com/justmarkham/DAT8/blob/master/data/chipotle.tsv

```python
import pandas as pd
```

```python
# 20. 데이터를 로드하라.

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df3 = pd.read_csv(DataUrl)
Ans = type(df3)
print(Ans)
```

```python
# 21. quantity컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하라

df3[df3.quantity==3].head(5)
```

```python
# 22. quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하라

df3.loc[df3['quantity']==3].head().reset_index(drop=True)
```

```python
# 23. quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라

df4 = df3[['quantity', 'item_price']]
df4
```

```python
df3
```

```python
# 24. item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라

df5 = df3
df5['new_price'] = df3['item_price'].str[1:].astype('float')
df5.head()
```

```python
# 25. new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라

df6 = df5[df5.new_price<=5]
len(df6)
```

```python
# 26. item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하라고 index 값을 초기화 하여라

df7 = df5[df5.item_name=='Chicken Salad Bowl'].reset_index(drop=True)
df7
```

```python
# 27. new_price값이 9 이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임을 추출하라

df8 = df5[(df5.new_price<=9) & (df5.item_name=='Chicken Salad Bowl')]
df8.head()
```

```python
# 28. df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화 하여라

df9 = df5.sort_values(by='new_price', ascending=True).reset_index(drop=True)
df9
```

```python
# 29. df의 item_name 컬럼 값중 Chips 포함하는 경우의 데이터를 출력하라

df10 = df5[df5.item_name.str.contains('Chips')]
df10
```

```python
# 30. df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라

df11 = df5.iloc[:, ::2]
df11
```

```python
# 31. df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하여라

df12 = df5.sort_values(by='new_price', ascending=False).reset_index(drop=True)
df12
```

```python
# 32. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱싱하라

df13 = df5[(df5.item_name == 'Steak Salad') | (df5.item_name == 'Bowl')]
# df13 = df.loc[(df.item_name =='Steak Salad') | (df.item_name =='Bowl')]
df13
```

```python
# 33. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

df14 = df5.loc[(df5.item_name =='Steak Salad') | (df5.item_name =='Bowl')].drop_duplicates('item_name', keep='first')
df14
```

```python
# 34. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라

df15 = df5.loc[(df5.item_name =='Steak Salad') | (df5.item_name =='Bowl')].drop_duplicates('item_name', keep='last')
df15
```

```python
# 35. df의 데이터 중 new_price값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱하라

new_price_avg = df5.new_price.mean()
df16 = df5[df5.new_price >= new_price_avg]
df16.head()
```

```python
# 36. df의 데이터 중 item_name의 값이 Izze 데이터를 Fizzy Lizzy로 수정하라

# df5.loc[df5.item_name =='Izze','item_name'] = 'Fizzy Lizzy'

df17 = df5.replace('Izze', 'Fizzy Izze')
df17
```
